```python
y = df.FloodProbability
df.drop('FloodProbability',axis=1,inplace=True)
X = df
X.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
def rf(xs, y, n_estimators=40, max_samples=200000,
       max_features=0.5, min_samples_leaf=5, **kwargs):
    return RandomForestRegressor(n_jobs=-1, n_estimators=n_estimators,
        max_samples=max_samples, max_features=max_features,
        min_samples_leaf=min_samples_leaf, oob_score=True).fit(xs, y)
```

```python
# m = rf(X_train, y_train);
```

```python
from sklearn.metrics import r2_score

def cal_score(m,X,y): 
    preds = m.predict(X)
    return r2_score(y,preds)
```

```python
# score = cal_score(m,X_test,y_test)
# score
```

```python
m2 = rf(X_train, y_train,n_estimators=80,max_samples=200000,
       max_features=0.4, min_samples_leaf=4);
score = cal_score(m2,X_test,y_test)
score
```
