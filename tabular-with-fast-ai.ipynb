```python
from fastai.tabular.all import *
```

```python
df = pd.read_csv('/kaggle/input/playground-series-s4e5/train.csv',index_col='id')
df.head()
```

```python
dls = TabularDataLoaders.from_df(df, y_names="FloodProbability",bs=1024)
```

```python
dls.show_batch()
```

```python
from sklearn.metrics import r2_score
learn = tabular_learner(dls, metrics=[accuracy,r2_score])
```

```python
learn.fine_tune(10)
```

```python
test_df = pd.read_csv('/kaggle/input/playground-series-s4e5/test.csv',index_col='id')
dl = learn.dls.test_dl(test_df)
```

```python
preds = learn.get_preds(dl=dl)
preds
```

```python
preds = preds[0].numpy()
```

```python
sub = pd.DataFrame(pd.Series(test_df.index),columns=['id'])
sub['FloodProbability'] = preds
sub.head()
```

```python
sub.to_csv("/kaggle/working/submission.csv",index=False)
```
